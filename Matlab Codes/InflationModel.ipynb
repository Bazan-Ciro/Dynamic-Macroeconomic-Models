## Model Assumptions

* <p style="text-align: justify;">The dynamic equilibrium condition of the money market is verified whenever, at an instant $t$, the real money supply, $L^s\left(M_t,P_t\right)$, equals the real money demand, $L^d\left(Y_t,i_t\right)$. That is:</p>
$$LM: L^s\left(M_t,P_t\right)=\frac{M_t}{P_t}=L^d\left(Y_t,i_t\right)=Y_t^{\psi}e^{-\theta{i_t}}\tag{1}$$
<p style="text-align: justify;"> Where $Y_t$ is the real output of domestic goods, $i_t$ represents the nominal interest rate, $M_t$ is the nominal stock of money, and $P_t$ represents the price index expressed in (money units/real consumption units). Furthermore, $\frac{M_t}{P_t}$ is the real money supply expressed in real consumption units, $\psi$ is the elasticity of real money demand with respect to real output, and $\theta$ is the semi-elasticity of real money demand with respect to the nominal interest rate. Where:</p>
  
  $$ \psi=\frac{Y_t}{L^d\left(Y_t,i_t\right)}\frac{\partial{L^d\left(Y_t,i_t\right)}}{\partial{Y_t}}=\frac{\partial\left[\ln{L^d\left(Y_t,i_t\right)}\right]}{\partial\left[\ln{\left(Y_t\right)}\right]}\tag{2} $$
  $$ \theta=-\frac{\partial\left[\ln{L^d\left(Y_t,i_t\right)}\right]}{\partial{i_t}}\tag{3} $$
  Applying logarithms to (1), we have:
  $$ \ln{\left(\frac{M_t}{P_t}\right)}=\ln{\left(Y_t^{\psi}e^{-\theta{i_t}}\right)}=\psi\ln{Y_t}-\theta{i_t} $$
  $$\ln{\left(M_t\right)}-\ln{\left(P_t\right)}=\psi\ln{\left(Y_t\right)}-\theta{i_t} \tag{4}$$
  Expressing the real production of domestic goods in logarithms gives the following results:
  $$y_t=\ln{\left(Y_t\right)}\tag{5}$$
  Then, the instantaneous growth rate of output will be:
  $$\dot{y_t}=\frac{\dot{Y_t}}{Y_t}   \tag{6}$$
  Defining the nominal monetary supply in logarithms gives the following:
  $$m_t=\ln{\left(M_t\right)}\tag{7}$$
  Then, the instantaneous growth rate of money will be:
  $$\dot{m_t}=\frac{\dot{M_t}}{M_t}   \tag{8}$$
  Similarly, defining the price index in logarithms, one obtains:
  $$p_t=\ln{\left(P_t\right)}\tag{9}$$
  <p style="text-align: justify;">Then the instantaneous growth rate of the national price index (inflation rate) will be:</p>
  $$\dot{p_t}=\frac{\dot{P_t}}{P_t}   \tag{10}$$
  <p style="text-align: justify;">Replacing equations (5), (7), and (9) in (4), the money market is characterised by a typical LM curve:</p>
  $$LM:m_t-p_t=\psi{y_t}-\theta{i_t} \tag{11} $$
  
 * <p style="text-align: justify;">The economy is populated by rational agents with perfect foresight. This assumption implies that the expected instantaneous growth rate of the price index, $\dot{p_t^e}$, is equal to the true value of the instantaneous growth rate of the price index, $\dot{p_t}$. That is, if:</p>

  \begin{equation}
 \left\{
\begin{aligned}
&p_t=lnP_t\\
&p_t^e=lnP_t^e
\end{aligned}
\right.
⇒ \left\{
\begin{aligned}
&\dot{p_t}=\frac{dlnP_t}{dt}=\frac{\dot{P_t}}{P_t}\\
&\dot{p_t}^e=\frac{dln{P_t}^e}{dt}=\frac{\dot{P_t^e}}{P_t^e}
\end{aligned}
\right.
⇒ \dot{p_t^e}=\dot{p_t}\\
\tag{12}
\end{equation}
* <p style="text-align: justify;">Inflation rate (instantaneous growth rate of the price index) is proportional to the gap between GDP and its potential value (both in logarithms):</p>
  $$\frac{\dot{P_t}}{P_t}=\mu\,ln\left(\frac{Y_t}{\bar{Y}_t}\right)=\mu[ln(Y_t)-ln(\bar{Y}_t)];\mu>0 \tag{13}    $$
  If the following is defined:
  $$ \bar{y}_t=ln\bar{Y}_t\tag{14}$$
  <p style="text-align: justify;">Then the instantaneous growth rate of the full employment level of output will be:</p>
  $$\dot{\bar{y}_t}=\frac{\dot{Y_t}}{Y_t}\tag{15}$$
  Therefore, replacing (5) and (14) in (13) gives the following:
  $$\dot{p_t}=\mu(y_t-\bar{y}_t)\tag{16}  $$
  <p style="text-align: justify;"> Where $\mu$ is the non-instantaneous speed of adjustment of the price index to the differences between the level of output and the level of output at full employment. Equation (16), a dynamic version of the Phillips curve, determines inflation and tells us that, if the level of economic activity is above the equilibrium level, $y_t>\bar{y}_t$, then prices will rise (demand-pull inflation: excess demand for goods and factors will drive up costs and prices).</p>
* <p style="text-align: justify;">Aggregate demand is assumed to be inversely affected by the nominal interest rate and directly affected by the expected rate of instantaneous growth of the price index (expected inflation).</p>
  $$ IS:{\ Y}_t^d=e^{\left[β_0-β_1\left(i_t-\frac{\dot{P_t^e}}{P_t^e}\right)\right]}\tag{17} $$
 <p style="text-align: justify;">Where $\beta_0$ is the autonomous component of aggregate demand (which includes government expenditure), $\beta_1$ is the semi-elasticity of aggregate demand for goods with respect to the nominal interest rate ($i_t$), $P_t^e$ the expected price index, $\dot{P_t^e}$ is the expected instantaneous rate of change of the price index, and $\frac{\dot{P_t^e}}{P_t^e}$ is the expected instantaneous growth rate of the price index (expected inflation). Where:</p>
 $$\beta_1=-\frac{\partial\left[ln{\left(Y_t^d\right)}\right]}{\partial i_t}\tag{18}$$
 Applying logarithms to (17) gives:
 $$\ln{\left(Y_t^d\right)}=ln\left[e^{\left[β_0-β_1\left(i_t-\frac{\dot{P_t^e}}{P_t^e}\right)\right]}\right]\tag{19} $$
 $$\ln{\left(Y_t^d\right)}=\beta_0-\beta_1\left(i_t-\frac{\dot{P_t^e}}{P_t^e}\right)\tag{20}$$
 Defining the following:
 $$y_t^d=\mathrm{ln}\left(Y_t^d\right)\tag{21}$$
 Then the instantaneous growth rate of aggregate demand for goods and services will be:
 $$\dot{y_t^d}=\frac{\dot{Y_t^e}}{Y_t^e}\tag{22}$$
 Replacing (12) and (21) in (20) gives:
 $$y_t^d=\beta_0-\beta_1\left(i_t-\dot{p_t^e}\right)\tag{23}$$
 <p style="text-align: justify;">Equation (23) is an IS curve in logarithmic terms in which the level of aggregate demand depends on its autonomous component, $\beta_0$, and the real interest rate, $r_t=i_t-\dot{p_t^e}$.</p>
 
 * <p style="text-align: justify;">Production is adjusted according to the following differential equation:</p>
 $$\frac{\dot{Y_t}}{Y_t}=φln\left(\frac{Y_t^d}{Y_t}\right)=φ\left[ln(Y_t^d)-ln(Y_t)\right]\tag{24}$$
 Replacing (6), (5) and (14) in (13) gives:
 $$\dot{y_t}=φ(y_t^d-y_t);φ>0\tag{25} $$
 <p style="text-align: justify;">Where $\varphi$ is the non-instantaneous speed of adjustment of output (in logarithmic terms) to excess demand or supply in the economy. Equation (25) tells us that output expressed in logarithmic terms (the growth rate of the economy) adjusts slowly to imbalances between aggregate demand and aggregate supply.</p>
 
 * All model parameters are positive.

## Obtaining the two-dimensional system of ODEs

<p style="text-align: justify;">In this model the exogenous variables are: $\beta_0, m_t,y_t$. While the endogenous variables are: $y_t, p_t,i_t,y_t$. To solve this model, following Torres (2012), $p_t$ and $y_t$ are considered as the endogenous variables of reference. Therefore, the differential equations for these variables are determinated below.</p>
<p style="text-align: justify;">To obtain the two differential equations that will determine the dynamic behaviour of the economy, the rest of the endogenous variables [the interest rate (obtained from (11) and the level of aggregate demand (23)] must be substituted into the adjustment equations of the reference endogenous variables [the price level (16) and the level of output (25)]. In the case of the differential equation for the price level, these variables do not appear, so that the first differential equation is exactly the same as the equation (16) provided by the model, i.e:</p>
 $$\dot{p_t}=\mu(y_t-\bar{y}_t)\tag{26}  $$
 <p style="text-align: justify;">To determine the differential equation of the product, the national nominal interest rate will first be cleared from equation (11):</p>
 $$i_t=\frac{1}{\theta}\left(\psi\,y_t-m_t+p_t\right)\tag{27}$$
 <p style="text-align: justify;">The aggregate demand of the national economy is obtained by substituting equations (12) and (27) into equation (23):</p>
 $$y_t^d=\beta_0-\beta_1\left[\frac{1}{θ}\left(ψy_t-m_t+p_t\right)-\dot{p_t}^e\right]=β_0-\beta_1\left[\frac{1}{θ}\left(ψy_t-m_t+p_t\right)-\dot{p_t}\right]\tag{28}$$
 Substituting (26) into (28) we obtain:
 $$y_t^d=\beta_0+\left(\mu\beta_1-\frac{\psi\beta_1}{\theta}\right)y_t-\frac{\beta_1}{\theta}p_t+\frac{\beta_1m_t}{\theta}-\beta_1\mu\bar{y}_t\tag{29}$$
 Replacing equation (29) in equation (25) gives:
 $$\dot{y_t}=-\left(\frac{φβ_1}{θ}\right)p_t+φ\left(μβ_1-\frac{\psiβ_1}{θ}-1\right)y_t+\left[φβ_0+\left(\frac{φβ_1}{θ}\right)m_t-\left(φμβ_1\right)\bar{y}_t\right]\tag{30}$$
 <p style="text-align: justify;">Consequently, the non-homogeneous linear autonomous system of differential equations defining the dynamic behaviour of this model, expressed in matrix form, is given by equations (26) and (30):</p>
  \begin{equation}
\begin{bmatrix}\dot{p_t} \\
    \dot{y_t}\\
\end{bmatrix} = \begin{bmatrix}0 & \mu \\
-\left(\frac{φβ_1}{θ}\right) & φ\left(μβ_1-\frac{\psiβ_1}{θ}-1\right)\end{bmatrix}. \begin{bmatrix}p_t \\
    y_t\\
\end{bmatrix}+\begin{bmatrix}0 & 0 & -\mu \\
φ & \frac{φβ_1}{θ} & -φμβ_1\end{bmatrix}. \begin{bmatrix}β_0 \\ m_t \\
    \bar{y}_t\\
\end{bmatrix}
\tag{31}\end{equation}
 
 
 \begin{equation}
\overbrace{\begin{bmatrix}\dot{p_t} \\
    \dot{y_t}\\
\end{bmatrix}}^{\dot{X_t}} = \overbrace{\begin{bmatrix}0 & \mu \\
-\left(\frac{φβ_1}{θ}\right) & φ\left(μβ_1-\frac{\psiβ_1}{θ}-1\right)\end{bmatrix}}^{A} . \overbrace{\begin{bmatrix}p_t \\
    y_t\\
\end{bmatrix}}^{X_t}+\overbrace{\begin{bmatrix}-μ\bar{y}_t \\
φβ_0+\left(\frac{φβ_1}{θ}\right)m_t-\left(φμβ_1\right)\bar{y}_t
\end{bmatrix}}^{b}
\tag{32}\end{equation}

<p style="text-align: justify;"> Since it has been assumed that all parameters of the model have a positive sign, the determinant of $A$ is strictly positive, $\begin{vmatrix}
A
\end{vmatrix}=\frac{μφβ_1}{θ}>0$. Hence, $A$ is invertible. Then, the equilibrium point of the system (32) can be calculated as follows:</p>

$$E=\left[\begin{matrix}p^E\\y^E\\\end{matrix}\right]=-A^{-1}{b}=\left[\begin{matrix}m_t+\left(\frac{θ}{β_1}\right)β_0-\left(ψ+\frac{θ}{β_1}\right)\bar{y}_t\\\bar{y}_t\\\end{matrix}\right]\tag{33}  $$

<p style="text-align: justify;">The eigenvalues $\lambda$ of the matrix $A$ are obtained from the following characteristic polynomial:</p>

$$p\left(\lambda\right)=\left|A-\lambda I\right|=\lambda^2-\left(trA\right)\lambda+\left|A\right|=0\ \ \ \ \ \ \tag{34}$$

<p style="text-align: justify;">Where $I$ is the identity matrix, $trA$ is the trace of $A$, and $\begin{vmatrix}
A
\end{vmatrix}$ is the determinant of $A$. Solving the characteristic polynomial gives the eigenvalues $\lambda_1$ and $\lambda_2$:</p>

$$\lambda_{1,2}=\frac{trA\pm\sqrt{\left(trA\right)^2-4\left|A\right|}}{2}=\frac{trA\pm\sqrt\Delta}{2}\ \ \ \ \tag{35}$$


<p style="text-align: justify;">The trace and the determinant of the matrix $A$, and the discriminant $Δ$ of the characteristic polynomial are given by:</p>


 \begin{equation}
 \left\{
\begin{aligned}
&trA=λ_1+λ_2=φ\left(μβ_1-\frac{ψβ_1}{θ}-1\right)\\
&\begin{vmatrix}
A
\end{vmatrix}=λ_1⋅λ_2=\frac{μφβ_1}{θ}>0\\
&Δ=trA^2-4\begin{vmatrix}
A
\end{vmatrix}=φ^2\left(μβ_1-\frac{ψβ_1}{θ}-1\right)^2-4\frac{μφβ_1}{θ}
\end{aligned}
\right.\tag{36}\end{equation}

## Analysis of a degenerate Hopf bifurcation 

<p style="text-align: justify;">In this section, in the neighbourhood of the equilibrium point of the linear dynamical system (32), possible qualitative changes in the dynamical nature of its phase curves as a consequence of the variation of some of its parameters are examined; that is, the possibility of the presence of a degenerate Hopf bifurcation is analyzed.</p>

<p style="text-align: justify;">$\textbf{Hopf Bifurcation theorem}$: Let $(x^E,y^E)$ be an equilibrium point of a two-dimensional autonomous system $\dot{x_t}=f(x_t,y_t,\mu)$, $\dot{y_t}=g(x_t,y_t,\mu)$ depending on some parameter $\mu \in \mathbf{R}$, being $f$ and $g$ smooth functions, and where $f(x^E,y^E,\mu)=0$ and $g(x^E,y^E,\mu)=0$. Let us assume that the Jacobian matrix of the system at the equilibrium point has complex eigenvalues $\lambda_{1,2}(\mu)=\alpha(\mu)\pm\beta(\mu) i$ for $\mu$ in a neighborhood of $\mu_0$. If the following criteria are verified:</p>

 <p style="text-align: justify;">(i)  $\alpha(\mu_0)=0$ and $\beta(\mu_0)=\beta_0>0$ (non-hyperbolicity criterion: purely pair of imaginary eigenvalues: $\lambda_{1,2}(\mu_0)=\pm\beta_0 i)$,</p>
 <p style="text-align: justify;">(ii)  $\frac{d\alpha(\mu)}{d\mu}\Big|_{\mu=\mu_0}=d\ne0$ (transversality criterion: the eigenvalues cross the imaginary axis with non-zero speed), and</p>
 <p style="text-align: justify;">(iii) $a\ne0$ [nondegeneracy condition: As Kuznetsov (2023) points out, this criterion implies that a certain combination of Taylor coefficients of the right-hand sides of the system (up to and including third-order coefficients) does not vanish], where</p>

 $$a=\frac{1}{16}[f_{x_tx_tx_t}+g_{x_tx_ty_t}+f_{x_ty_ty_t}+g_{y_ty_ty_t}]+\frac{1}{16\beta_0}[f_{x_ty_t}(f_{x_tx_t}+f_{y_ty_t})-g_{x_ty_t}(g_{x_tx_t}+g_{y_ty_t})-f_{x_tx_t}g_{x_tx_t}+f_{y_ty_t}g_{y_ty_t}],$$ and all partial derivatives are evaluated at the bifurcation point $(x^E,y^E,\mu_0).$

 <p style="text-align: justify;">"Then a unique curve of periodic solutions emanates (bifurcates) from
$(x^E,y^E)$ into the region $\mu>\mu_0$ (for $ad<0$), or for $\mu<\mu_0$ (for $ad>0$), and the equilibrium point changes stability as the parameter value passes through $\mu_0$, with the newly formed periodic
orbits having the same stability as that of the equilibrium on the other side of $\mu_0$. The Hopf bifurcation is called supercritical if the equilibrium goes from stable to unstable and the new periodic orbits are stable. In contrast, if the equilibrium changes form unstable to stable and the periodic orbits are unstable, the Hopf bifurcation is called subcritical", (Brown, 2018).</p>

<p style="text-align: justify;">$\textbf{Applying the Hopf theorem}$: Next, using Hopf's theorem, the conditions under which a Hopf bifurcation occurs in the model are analyzed, considering $\mu$ as the bifurcation parameter.</p>  

<p style="text-align: justify;">The Jacobian associated with (32), evaluated at the equilibrium point given by (33), and in which $\mu$ is treated as the only active parameter (the rest of the parameters could be equal to one), is given by:</p>  
    
$$A\left(\mu\right)=\left[\begin{matrix}0&\mu\\-\left(\frac{\varphi}{\mu\theta-\psi}\right)&\varphi\left(\mu\beta_1-\frac{\psi}{\mu\theta-\psi}-1\right)\\\end{matrix}\right]\tag{37}$$

<p style="text-align: justify;">Considering (36), to obtain complex conjugate eigenvalues with zero real part and periodic trajectories, the following conditions must be verified:</p>  

 \begin{equation}
 \left\{
\begin{aligned}
&trA(\mu)=φ\left(μβ_1-\frac{ψβ_1}{θ}-1\right)=0\Leftrightarrow\ \mu\beta_1-\frac{\psi\beta_1}{\theta}-1=0 \left(\mu=\frac{\psi\beta_1+\theta}{\theta\beta_1}=\mu_0\right)\\
&\begin{vmatrix}
A(\mu)
\end{vmatrix}=\frac{μφβ_1}{θ}>0\\
&Δ=-4\frac{μφβ_1}{θ}<0
\end{aligned}
\right.\tag{38}\end{equation}

Hence, based on (35) and (38), it follows that:

$$\lambda_j\left(\mu\right)=\alpha\left(\mu\right)\pm\beta\left(\mu\right)i\in\mathbb{C}^2,\ j=\left\{1,2\right\}\ \ \ \ (39)$$

<p style="text-align: justify;">Where $\alpha\left(\mu\right)=\frac{trA(\mu)}{2}=\frac{\varphi\left(\mu\beta_1-\frac{\psi\beta_1}{\theta}-1\right)}{2}$ and $\beta\left(\mu\right)=\frac{\sqrt{-\Delta}}{2}=\sqrt{\frac{\mu\varphi\beta_1}{\theta}}$. Likewise, we see that for $\mu=\mu_0$, we have $\alpha\left(\mu_0\right)=0$, and $\beta\left(\mu_0\right)=\sqrt{\frac{\mu_0\varphi\beta_1}{\theta}}=\beta_0>0$, satisfying criterion (i). Also  $\frac{d\alpha(\mu)}{d\mu}\Big|_{\mu=\mu_0}=\frac{\psi\beta_1}{2}=d>0$, satisfying criterion (ii). And finally, with $f(p_t,y_t,\mu_0)=\mu_0({y}_t-\bar{y}_t)$ and $g(p_t,y_t,\mu_0)=-\left(\frac{\varphi\beta_1}{\theta}\right){p}_t+\varphi\left(\mu_0\beta_1-\frac{\psi\beta_1}{\theta}-1\right){y}_t+(\varphi\beta_0+\frac{\varphi\beta_1}{\theta}m_t-\varphi\mu\beta_1\bar{y}_t)$, we have that:</p>

$$f_{p_{t}}(p^E,y^E,\mu_0)=f_{p_{t}p_{t}}(p^E,y^E,\mu_0)=f_{p_{t}p_{t}p_{t}}(p^E,y^E,\mu_0)=f_{p_{t}y_{t}}(p^E,y^E,\mu_0)=f_{p_{t}y_{t}y_{t}}(p^E,y^E,\mu_0)=0$$

$$g_{p_{t}}(p^E,y^E,\mu_0)=-\left(\frac{\varphi\beta_1}{\theta}\right)$$

$$g_{y_{t}}(p^E,y^E,\mu_0)=\varphi\left(\mu_0\beta_1-\frac{\psi\beta_1}{\theta}-1\right)$$

$$g_{p_{t}p_{t}}(p^E,y^E,\mu_0)=g_{p_{t}y_{t}}(p^E,y^E,\mu_0)=g_{p_{t}p_{t}y_{t}}(p^E,y^E,\mu_0)=g_{y_{t}y_{t}y_{t}}(p^E,y^E,\mu_0)=0$$

<p style="text-align: justify;">Thus, since all relevant partial derivatives are zero, it follows that $a=0$, which violates criterion (iii). Then, no periodic solutions exist for $\mu ≠ \mu_0$, and the model exhibits a degenerate Hop bifurcation. However, system (32) has infinitely many periodic orbits for $\mu = \mu_0$ and its equilibrium point is a center.</p>

## Simulating the dynamic inflation model with Python

### Import Python Libraries

````python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
import numpy as np # library for scientific computing 
from scipy.integrate import ode # Library containing numerical routines for solving ordinary differential equations (ODEs)
import matplotlib.pyplot as pl # Library for generating 2D graphics
````

### Establish the system of ODEs that describe the dynamics of the inflation model

````python
def dxdt(t, x, a, b, c, d, e, f):
    y, p = x
    return np.array([
        a * y + b * p + c,
        d * y + e * p + f
    ])
````

### Define the parameters of the inflation model

````python
ybar = 4010
phi = 0.9
betha0 = 1774
theta = 0.9
betha1 = 119.076923
psi = 0.01
m = 58

mu0 = (psi * betha1 + theta) / (theta * betha1)
v = 0.0196 - mu0
mu_values = [0.01, mu0, mu0 + v]
````

### Time settings and initial value

````python
t_0 = 0.0
t_max = 300
N_spacing_in_t = 5000
t = np.linspace(t_0, t_max, N_spacing_in_t)
x0 = np.array([3957.71, 1.5])  # [y_0,p_0]
````

### Loop over each $\mu$ value

````python
for idx, mu in enumerate(mu_values):
    # Auxiliar parameters
    a = phi * (mu * betha1 - psi * betha1 / theta - 1)
    b = -phi * betha1 / theta
    c = phi * betha0 + phi * betha1 / theta * m - phi * mu * betha1 * ybar
    d = mu
    e = 0
    f = -mu * ybar
    pE = m + theta * betha0 / betha1 - (psi + theta / betha1) * ybar
    yE = ybar

    # Solve the ODE system
    r = ode(dxdt).set_integrator('dopri5')
    r.set_initial_value(x0, t_0)
    r.set_f_params(a, b, c, d, e, f)

    x_result = np.zeros((len(t), 2))
    for it, t_ in enumerate(t):
        x = r.integrate(t_)
        x_result[it, :] = x

    production = x_result[:, 0]
    price = x_result[:, 1]

    # Figure 1: Phase portrait
    pl.figure(figsize=(7, 6))
    pl.plot(production, price, '-', lw=1)
    pl.plot(x0[0], x0[1], 'ok', markersize=4)
    pl.plot(yE, pE, 'ok', markersize=4)
    pl.annotate('I', xy=(x0[0], x0[1]), xytext=(x0[0]+2, x0[1] - 0.02))
    pl.annotate('E', xy=(yE, pE), xytext=(yE+2, pE - 0.02))
    pl.xlabel('$y$')
    pl.ylabel('$p$', rotation=0)
    pl.title(f'Phase Portrait (mu = {mu:.5f})')

    # Vector field
    y_min, y_max = min(production.min(), yE) - 10, max(production.max(), yE) + 10
    p_min, p_max = min(price.min(), pE) - 0.1, max(price.max(), pE) + 0.1
    Y, P = np.meshgrid(np.linspace(y_min, y_max, 25), np.linspace(p_min, p_max, 25))
    U, V = np.zeros_like(Y), np.zeros_like(P)
    scale_y = 1
    scale_p = 35

    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            if abs(Y[i, j] - yE) < 1e-3 and abs(P[i, j] - pE) < 1e-3:
                continue
            vec = dxdt(0, [Y[i, j], P[i, j]], a, b, c, d, e, f)
            dy, dp = vec[0] * scale_y, vec[1] * scale_p
            norm = np.sqrt(dy**2 + dp**2)
            if norm != 0:
                U[i, j] = dy / norm
                V[i, j] = dp / norm

    pl.quiver(Y, P, U, V, color='r', alpha=0.6)
    pl.legend()
    pl.tight_layout()

    # Figure 2: Time paths
    pl.figure(figsize=(8, 6))
    pl.subplot(2, 1, 1)
    pl.plot(t, production, 'b')
    pl.xlabel('$t$')
    pl.ylabel('$y$', rotation=0)
    pl.title(f'Time Path of Production (mu = {mu:.5f})')

    pl.subplot(2, 1, 2)
    pl.plot(t, price, 'b')
    pl.xlabel('$t$')
    pl.ylabel('$p$', rotation=0)
    pl.title(f'Time Path of Price Index (mu = {mu:.5f})')
    pl.tight_layout()

pl.show()
````

# Summary of ideas

<p style="text-align: justify;">In this notebook, it was shown that the system describing the dynamics of the model is structurally unstable, since the Jacobian matrix $A\left(\mu\right)$ of the system, given by (37), can exhibit eigenvalues with a zero real part (marginally stable center). Furthermore, Using Hopf's theorem, it was verified that the model experiences a degenerate Hopf bifurcation for the critical value of the price index adjustment rate, $\mu_0$, given by the first line of (38).</p>

<p style="text-align: justify;">Finally, running numerical simulations, it was found that the qualitative properties and dynamic stability of the phase trajectories of the economy's price level and GDP change when going from values of $\mu<\mu_0$ (spiral converging asymptotically and oscillatingly to the steady state) to $\mu=\mu_0$ (marginally stable centre: closed, periodic, constant-amplitude orbit) and to values of $\mu>\mu_0$ (spiral diverging from the steady state in an oscillating manner), ceteris paribus.</p>
