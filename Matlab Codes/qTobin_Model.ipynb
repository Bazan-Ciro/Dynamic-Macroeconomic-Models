## Introduction

According to Romer (2018), the study of investment is important because:

* <p style="text-align: justify;">The percent of output that goes to investment depends on the interaction between the investment demand of firms and the savings supply of households. Investment demand is therefore one of the factors influencing the long-run growth of an economy.</p>

* <p style="text-align: justify;">Investment is the most volatile component of GDP in the short run; therefore, investment demand may be important in explaining economic fluctuations in the short run.</p>

<p style="text-align: justify;">
    Tobin's model with adjustment costs of capital stock provides a theoretical framework for the analysis of the firm's investment decisions in a more realistic way than the basic neoclassical model (where the adjustment of capital to sudden discrete changes, e.g. in the interest rate, would have to be made at a very high rate). In addition, Tobin's model provides a link between investment and the shadow value of capital in the firm, $q$.
</p>

## Model Assumptions

* There are $N$ identical firms in the industry.

* <p style="text-align: justify;">If setup and capital purchase costs were irrelevant (we will later lift this initial assumption), the real profits of a representative firm at $t$ would be proportional to its capital stock, $k(t)$, and decreasing with the industry-wide capital stock, $K(t)$. That is:</p>

  $$  \pi\lbrackùêæ(ùë°)\rbrackùëò(ùë°)\tag{1} $$

  
  where:


\begin{equation}
\left\{
\begin{aligned}
&K(t) = Nk(t) \\
&\pi'\left[K(t)\right]< 0
\end{aligned}
\right.
\tag{2}
\end{equation}

* <p style="text-align: justify;">The production function has constant returns to scale. This assumption implies that the profits of the representative firm can be represented in the multiplicative manner shown in equation (1).</p>
* Product markets are competitive.
* The supply of factors of production, except capital, is perfectly elastic.
* The demand curve for the product of this industry is negatively sloped.
* The firm faces adjustment costs to its capital stock.
* <p style="text-align: justify;">Adjustment costs are convex with respect to the rate of change in the firm's capital stock.</p>
* Specifically, the adjustment costs, $ {C}(\dot{k})$, satisfy:
\begin{equation}
\left\{
\begin{aligned}
&\mathcal{C}(0) = 0 \\
&\mathcal{C}'(0) = 0 \\
&\mathcal{C}''(\dot{k}) > 0
\end{aligned}
\right.
\tag{3}
\end{equation}
<p style="text-align: justify;">This assumption implies that the firm incurs positive adjustment costs when its capital stock changes either up or down, and that the marginal adjustment cost is increasing in the size of the adjustment (in the instantaneous rate of change of the capital stock).</p>

* <p style="text-align: justify;">The purchase price of capital goods is constant equal to $1$; then, there are only internal adjustment costs (e.g. the installation of new capital or the training of workers in the operation of new machinery).</p>

* There is no capital depreciation. Therefore, it follows that:

  $$ {C}(\dot{k})=ùêº(ùë°)\tag{4} $$

  Where $I$ is the firm's investment.

* <p style="text-align: justify;">Removing the assumption of irrelevant installation and capital purchase costs: To remove this initial assumption, new capital purchase costs (investment) and new capital installation costs (adjustment costs) will have to be discounted from equation (1). That is:</p>

  $$ ùúã[ùêæ(ùë°)]‚ãÖùëò(ùë°)‚àíùêº(ùë°)‚àíùê∂[ùêº(ùë°)]\tag{5} $$

  <p style="text-align: justify;">By investing, the firm incurs a sunk cost of ‚Äòinstallation‚Äô over and above the cost of the investment itself.</p>

* The firm maximises the present value of its profits:

$$  \Pi = \int_{t=0}^{+\infty} e^{-rt} \left[ \pi[K(t)] \cdot k(t) - I(t) - \mathcal{C}[I(t)] \right] dt \tag{6} $$
 $ \quad \text{s.t.}$
$$ K(0) = K_0 > 0 \quad \text{(given)} $$
$$\dot{k}(t) = I(t) \quad \text{(state equation)} $$
$$\lim_{t \to +\infty} \left[ k(t) \cdot q(t) e^{-rt} \right] = 0 \quad \text{(transversality condition)} $$

<p style="text-align: justify;">Where the real interest rate $r$ is assumed to be constant, and each firm takes the industry-wide capital stock path, $K(t)$, as given, and chooses its investment over time to maximise (6) given the path of $K(t)$. Intuitively, the transversality condition tells us that it cannot be optimal for a company to hold valuable capital indefinitely.</p>

## Solving the Model

<p style="text-align: justify;">To solve problem (6) we are going to use the dynamic optimisation technique called optimal control. To solve (6), we will first define a function analogous to the Lagrangian, called the Hamiltonian. That is:</p>
$$ ùêª[ùëò(ùë°),ùêº(ùë°)]=e^{-rt} [ùúã[ùêæ(ùë°)]‚ãÖùëò(ùë°)‚àíùêº(ùë°)‚àíùê∂[ùêº(ùë°)]]+ùúÜ(ùë°)ùêº(ùë°) \tag{7} $$
<p style="text-align: justify;">Where $Œª(t)$, gives the marginal impact of an exogenous one-unit increase in $k(t)$ on the value of the firm's lifetime profits (from $0$ to $+‚àû$) and discounted to the instant $0$.  Then, if we define:</p> 
$$ ùëû(ùë°)=ùëí^{rt} ùúÜ(ùë°)‚áíùúÜ(ùë°)=ùëí^{-rt} ùëû(ùë°)\tag{8}  $$
Moreover, if analogously to (8) we define:
$$ \hat{H}¬†ÃÇ[ùëò(ùë°),I(ùë°)]=ùëí^{rt} ùêª[ùëò(ùë°),I(ùë°)]‚áíùêª[ùëò(ùë°),I(ùë°)]=ùëí^{-rt} \hat{H}[ùëò(ùë°),I(ùë°)]¬†ÃÇ\tag{9}$$
Substituting (8) into (7) we obtain:
$$ ùêª[ùëò(ùë°),ùêº(ùë°)]=ùëí^{-rt} [ùúã[ùêæ(ùë°)]‚ãÖùëò(ùë°)‚àíùêº(ùë°)‚àíùê∂[ùêº(ùë°)]]+ùëû(ùë°) ùëí^{-rt} ùêº(ùë°) \tag{10}    $$
Now, replacing (9) in (10) we obtain the Hamiltonian at present value. That is:
$$ \hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]=ùúã[ùêæ(ùë°)]‚ãÖùëò(ùë°)‚àíùêº(ùë°)‚àíùê∂[ùêº(ùë°)]+ùëû(ùë°)ùêº(ùë°)  \tag{11}   $$
<p style="text-align: justify;">Where, $I(t)$, the variable that can be freely controlled and can change instantaneously, is called the control variable; $k(t)$, the variable whose value at any instant $t$ is determined by decisions taken in the past, is called the state variable; and the shadow value of the state variable, $q(t)$, Tobin's $q$, is called the co-state variable, and is the value to the firm of an additional unit of capital at time $t$ in monetary units at time $t$.</p>
The first-order optimality conditions derived from the Hamiltonian at present value are given by:

$$ \frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïùêº(ùë°)}= 0\tag{12} $$ 
$$ \dot{k}(t) =\frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïq(ùë°)} \tag{13} $$
$$ \dot{q}(t) =-\frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïk(ùë°)}+rq(t) \tag{14} $$

For our model, the first order optimality conditions are:
$$ \frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïùêº(ùë°)}=‚àí1‚àíùê∂^{\,‚Ä≤} [ùêº(ùë°)]+ùëû(ùë°)= 0\tag{15} $$ 
$$ \dot{k}(t) =\frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïq(ùë°)}=ùêº(ùë°) \tag{16} $$
$$ \dot{q}(t) =-\frac{ùúï\hat{H}¬†ÃÇ[ùëò(ùë°),ùêº(ùë°)]}{ùúïk(ùë°)}+rq(t)=‚àíùúã[ùêæ(ùë°)]+ùëüùëû(ùë°) \tag{17} $$
<p style="text-align: justify;">Equations (15), (16) and (17) together with the transversality condition in (6) characterise the optimal decisions of the firm.</p>
From (15) it results:
$$ ùëû(ùë°)=1+ùê∂^{\,‚Ä≤} [ùêº(ùë°)] \tag{18} $$
<p style="text-align: justify;">Equation (18) tells us that the cost of acquiring a unit of capital, $1+C'[I(t)]$, is equal to the purchase price of capital goods (which is fixed at $1$) plus the marginal cost of adjusting capital. Then, (18) states that the firm invests up to the point where the cost of acquiring capital, $1+ùê∂^{\,‚Ä≤}[I(t)]$, equals the value of capital, $q(t)$. From (18) it follows:</p>
$$ ùê∂^{\,‚Ä≤} [ùêº(ùë°)]=ùëû(ùë°)‚àí1‚áíùê∂^{{\,‚Ä≤}{‚àí1}} ‚ü®ùê∂^{\,‚Ä≤} [ùêº(ùë°)]‚ü©=ùê∂^{{\,‚Ä≤}{‚àí1}}[ùëû(ùë°)‚àí1] $$
$$ ùêº(ùë°)=ùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1]\tag{19}$$
Now, from (16), we have that:
$$ùêº(ùë°)=\dot{k}(t) \tag{20}  $$
Replacing (19) in (20), it follows that:
$$ùêº(ùë°)=\dot{k}(t)=ùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1] \tag{21}  $$
<p style="text-align: justify;">On the other hand, given that we have assumed that the $N$ firms in the industry are identical, and that therefore $q$ is identical for all of them, then all firms choose the same value of $I$. Therefore, the rate of change of the aggregate capital stock, is given by the product between the number of firms in the industry and the value of $I$ satisfying (21). That is to say:</p>
$$\dot{K}(t)=ùëÅ \dot{k}(t)=ùëÅùêº(ùë°)=ùëÅùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1]=ùëî[ùëû(ùë°)] \tag{22}  $$
Now, we will assume that:

\begin{equation}
\left\{
\begin{aligned}
&ùëî(1)=0 \\
&ùëî^‚Ä≤ [ùëû(ùë°)]>0 \\
\end{aligned}
\right.
\tag{23}
\end{equation}
Considering that we have assumed that:
\begin{equation}
\left\{
\begin{aligned}
&ùê∂^{\,‚Ä≤}(ùêº)>0\\
&ùëî^‚Ä≤[ùëû(ùë°)]>0\\
&ùëî(1)=0\\
&ùê∂^‚Ä≤(0)=0\\
\end{aligned}
\right.
\tag{24}
\end{equation}
It follows from (22) and (24) that:
\begin{equation}
\left\{
\begin{aligned}
&\dot{K}(t)>0\Leftrightarrow q>1\\
&\dot{K}(t)=0\Leftrightarrow q=1\\
&\dot{K}(t)<0\Leftrightarrow q<1\\
\end{aligned}
\right.
\tag{25}
\end{equation}
On the other hand, from (17), we have that:
$$ ùúã[ùêæ(ùë°)]=ùëüùëû(ùë°)‚àí\dot{q}(t)     \tag{26}$$
<p style="text-align: justify;">Equation (26) tells us that the marginal revenue of the firm's capital, $ùúã[ùêæ(ùë°)]$, is equal to its user cost, $ùëüùëû(ùë°)‚àí\dot{q}(t)$. Now, from (26), it follows:</p>
$$\dot{q}(t) = ùëüùëû(ùë°)‚àí ùúã[ùêæ(ùë°)]   \tag{27}$$
<p style="text-align: justify;">In summary, the differential equations describing the dynamics of the model are equations (22) and (27). That is:</p>
\begin{equation}
\left\{
\begin{aligned}
&\dot{K}(t)=ùëÅùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1]=ùëî[ùëû(ùë°)]\\
&\dot{q}(t) = ùëüùëû(ùë°)‚àí ùúã[ùêæ(ùë°)]\\
\end{aligned}
\right.
\tag{28}
\end{equation}
<p style="text-align: justify;">Now, using equation (28) we will construct the phase diagram of the model, in the phase plane $(K, q)$, to analyse its dynamic qualities in a qualitative way. To do so, we will first determine the ceroclines of the model. To determine the nullclines, we equal to zero the two differential equations given by (28). That is:</p>
\begin{equation}
\left\{
\begin{aligned}
&\dot{K}(t)=ùëÅùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1]=ùëî[ùëû(ùë°)]=0\\
&\dot{q}(t) = ùëüùëû(ùë°)‚àí ùúã[ùêæ(ùë°)]=0\\
\end{aligned}
\right.
\tag{29}
\end{equation}
Taking into account the first expression of (23) and (25), it follows that:
$$ \dot{K}(t)=ùëÅùê∂^{{\,‚Ä≤}{‚àí1}} [ùëû(ùë°)‚àí1]=ùëî[ùëû(ùë°)]=0\Leftrightarrowùëû(ùë°)=1\,‚àÄùë° \tag{30}$$
<p style="text-align: justify;">Equation (30) will be a horizontal line cutting the vertical axis of the phase plane $(K, q)$ at $q = 1$, since it does not depend on $K$.</p>
<p style="text-align: justify;">The second equation of (29) implies that over the nullcline $\dot{q}(t)=0$ it is satisfied that $q$ is constant in time. However, in the phase plane, all points on the nullcline have values of $q$ decreasing in $K$, since $œÄ[K(t)]$ is decreasing in $K$.</p>
$$\dot{q}(t)=ùëüùëû(ùë°)‚àíùúã[ùêæ(ùë°)]=0‚â°[ùëû=\frac{ùúã[ùêæ(ùë°)]}{ùëü}=ùëêùë°ùëí\, ‚àÄùë°‚àà(0,+‚àû)]\tag{31}$$

## Interpretations of Tobin's q

Tobin's $q$ can be interpreted in the following ways:

* <p style="text-align: justify;">It is the shadow price of installed capital, $k(t)$. That is, the shadow value of the state variable.</p>
* <p style="text-align: justify;">It gives the marginal impact of an exogenous one-unit increase in $k(t)$ on the value of the firm's profits earned from $t$ to $+‚àû$ and discounted to instant $t$.</p>
* <p style="text-align: justify;">It is the value to the firm of an additional unit of capital at time $t$ in monetary units at time $t$.</p>
* <p style="text-align: justify;">It is the discounted value at time $t$ of the future marginal revenue of an additional unit of capital. That is:</p>
  $$ùëû(ùë°)=\int_{ùúè=ùë°}^{+\infty}\left[ùúã[ùêæ(ùúè)] ùëí^{‚àíùëü(ùúè‚àíùë°) }\right] ùëëùúè   \tag{32}  $$
   <p style="text-align: justify;">Now, we will show how to obtain equation (32). Multiplying (26) by the intertemporal discount factor yields:</p>
  $$ùúã[ùêæ(ùë°)] ùëí^{‚àíùëüùë°}=‚àí\left[\dot{q}(t)‚àíùëüùëû(ùë°)\right] ùëí^{‚àíùëüùë°}=‚àí\frac{ùëë\left[ùëû(ùë°) ùëí^{‚àíùëüùë°}\right ]}{ùëëùë°} \tag{33}$$
  Integrating (33) and discounting to period $t$ the future flows, we obtain $q(t)$:
  $$\int_{ùúè=ùë°}^{+\infty}\left[ùúã[ùêæ(ùúè)] ùëí^{‚àíùëü(ùúè‚àíùë°) }\right] ùëëùúè=-\int_{ùúè=ùë°}^{+\infty}\frac{ùëë\left[ùëû(ùúè) ùëí^{‚àíùëü(ùúè‚àíùë°) } \right]}{ùëëùë°} ùëëùúè  $$

  $$ \int_{ùúè=ùë°}^{+\infty}\left[ùúã\left[ùêæ(ùúè)\right] ùëí^{‚àíùëü(ùúè‚àíùë°) }\right] ùëëùúè = -\left[ q(ùúè) \, ùëí^{‚àíùëü(ùúè‚àíùë°)} \right]_{ùúè=ùë°}^{+\infty}=q(t) - \lim_{ùúè \to +‚àû} \left[\frac{ùëû(ùúè)}{ùëí^{ùëü(ùúè‚àíùë°)}}\right] $$
  <p style="text-align: justify;">But, if we assume that it is fulfilled that $\frac{\frac{ùëëùëû(ùúè)}{ùëëùúè}}{ùëû(ùúè)}<\frac{\frac{ùëëùëí^{ùëü(ùúè‚àíùë°)}}{ùëëùúè}}{ùëí^{ùëü(ùúè‚àíùë°)}}=r$, then $\lim_{ùúè \to +‚àû} \left[\frac{ùëû(ùúè)}{ùëí^{ùëü(ùúè‚àíùë°)}}\right]=0$. Finally, substituting this limit in the previous expression gives the following result:</p>
  $$ùëû(ùë°)=\int_{ùúè=ùë°}^{+\infty}\left[ùúã\left[ùêæ(ùúè)\right] ùëí^{‚àíùëü(ùúè‚àíùë°) }\right] ùëëùúè   \tag{34}  $$

## Quantitative Analysis of a Particular Case

<p style="text-align: justify;">To numerically solve, using Matlab software, the system of Ordinary Differential Equations (ODEs) that govern the dynamics of the model, it will be assumed that: </p>

&ùë°ùëüùê¥=ùëü>0\\
&|ùê¥|=‚àíùëè\frac{N}{ùõº}<0\\
&Œî=(ùë°ùëüùê¥)^2‚àí4|ùê¥|=ùëü^2+4 ùëè\frac{N}{ùõº}>0\\
\end{aligned}
\right.
\tag{41}
\end{equation}
<p style="text-align: justify;">For this particular case, the system of ODEs given by (40) is linear due to the assumptions adopted in (35). By (41) and/or (42), the steady state of (40) is a saddle point. The eigenvalues associated with the matrix $A$ shall be:</p>
\begin{equation}
\left\{
\begin{aligned}
&\lambda_1=\frac{ùë°ùëüùê¥+\sqrt{Œî}}{2}=\frac{r+\sqrt{ùëü^2+4 ùëè\frac{N}{ùõº}}}{2}>0\\
&\lambda_2=\frac{ùë°ùëüùê¥-\sqrt{Œî}}{2}=\frac{r-\sqrt{ùëü^2+4 ùëè\frac{N}{ùõº}}}{2}<0\\
\end{aligned}
\right.
\tag{42}
\end{equation}
The eigenvectors of the matrix $A$ of the system (40) are given by:
\begin{equation}
V_1=\begin{bmatrix}1 \\
\frac{\alpha\lambda_1}{N}
\end{bmatrix}
\tag{43}\end{equation}
\begin{equation}
V_2=\begin{bmatrix}1 \\
\frac{\alpha\lambda_2}{N}
\end{bmatrix}
\tag{44}\end{equation}
The steady state of the system (40) is given by:
\begin{equation}
\overbrace{\begin{bmatrix}K^* \\
    q^*\\
\end{bmatrix}}^{X^*} = \overbrace{(\frac{-\alpha}{bN})\begin{bmatrix}r & -\frac{N}{\alpha} \\
-b & 0\end{bmatrix}}^{A^{-1}} \times \overbrace{\begin{bmatrix}-\frac{N}{\alpha} \\
-a
\end{bmatrix}}^{b}=\begin{bmatrix}\frac{a-r}{b} \\
1
\end{bmatrix}
\tag{45}\end{equation}
The solution of the system (40) is given by:
$$ùëã=ùê¥_1 ùëâ_1 ùëí^{\lambda_1 ùë°}+ùê¥_2 ùëâ_2 ùëí^{\lambda_2 ùë°}+ùëã^‚àó\tag{46}$$
Substituting (43), (44) and (45) into (46), it follows:
$$\begin{bmatrix}K(t) \\
    q(t)\\
\end{bmatrix}=ùê¥_1\begin{bmatrix}1 \\
\frac{\alpha\lambda_1}{N}
\end{bmatrix}ùëí^{\lambda_1 ùë°}+ùê¥_2\begin{bmatrix}1 \\
\frac{\alpha\lambda_2}{N}
\end{bmatrix}ùëí^{\lambda_2 ùë°}+\begin{bmatrix}K^* \\
    q^*\\
\end{bmatrix}$$
\begin{equation}
\left\{
\begin{aligned}
&K(t)=A_1ùëí^{\lambda_1 ùë°}+A_2ùëí^{\lambda_2 ùë°}+K^*\\
&q(t)=\frac{\alpha\lambda_1A_1}{N}ùëí^{\lambda_1 ùë°}+\frac{\alpha\lambda_2A_2}{N}ùëí^{\lambda_2 ùë°}+q^*\\
\end{aligned}
\right.
\tag{47}
\end{equation}
 <p style="text-align: justify;">Since the eigenvalue $\lambda_1 > 0$, the exponential term $ùëí^{\lambda_1 ùë°}$ of the equations given by (47) will explode to more or less infinity as $t$ tends to infinity (depending on the sign of $A_1$). In order to reach long-run stationary equilibrium (along the Stable branch: $SB$) it will be necessary to eliminate this explosive effect. To do this we must make $A_1$ equal to zero. Substituting $A_1=0$ in (47) we will have the equations of $K(t)$ and $q(t)$ along the $SB$ of the saddle. That is:</p>
\begin{equation}
\left\{
\begin{aligned}
&K(t)=A_2ùëí^{\lambda_2 ùë°}+K^*\\
&q(t)=\frac{\alpha\lambda_2A_2}{N}ùëí^{\lambda_2 ùë°}+q^*\\
\end{aligned}
\right.
\tag{48}
\end{equation}
Taking into account (6), and evaluating the first expression of (48) at $t = 0$, we have that:
$$ùêæ(0)=ùê¥_2+ùêæ^‚àó=ùêæ_0‚áíùê¥_2=ùêæ_0‚àíùêæ^‚àó\tag{49}$$
<p style="text-align: justify;">Replacing (49) in (48), we finally have the solution of the system (40) (along the $BE$) as a function of known parameters:</p>
\begin{equation}
\left\{
\begin{aligned}
&K(t)=(ùêæ_0‚àíùêæ^‚àó)ùëí^{\lambda_2 ùë°}+K^*\\
&q(t)=\frac{\alpha\lambda_2(ùêæ_0‚àíùêæ^‚àó)}{N}ùëí^{\lambda_2 ùë°}+q^*\\
\end{aligned}
\right.
\tag{50}
\end{equation}
 <p style="text-align: justify;">Substituting the second expression of (48) in (38), and taking into account that by (45) $q^* = 1$, we have that the optimal investment of the company at each instant $t$ will be:</p>
$$ùêº(ùë°)=\frac{1}{ùõº} \left[\frac{ùõºùúÜ_2 (ùêæ_0‚àíùêæ^‚àó )}{ùëÅ} ùëí^{\lambda_2 ùë°}+ùëû^‚àó‚àí1\right]=\frac{ùúÜ_2 (ùêæ_0‚àíùêæ^‚àó )}{ùëÅ} ùëí^{\lambda_2 ùë°} \tag{51}  $$
<p style="text-align: justify;">Replacing the first expression of (50) and (51) in the expressions of (35) respectively, we have:</p>
\begin{equation}
\left\{
\begin{aligned}
&ùúã(ùë°)=(ùëé‚àíùëèùêæ^‚àó )‚àíùëè(ùêæ_0‚àíùêæ^‚àó ) ùëí^{\lambda_2 ùë°}\\
&ùê∂(ùë°)=\frac{ùõº\lambda_2^2 (ùêæ_0‚àíùêæ^‚àó )^2}{2ùëÅ^2} ùëí^{\lambda_2 ùë°}\\
\end{aligned}
\right.
\tag{52}
\end{equation}
Substituting (51) in (20), we have that:
$$\dot{k}(t)=\frac{\lambda_2 (ùêæ_0‚àíùêæ^‚àó )}{ùëÅ} ùëí^{\lambda_2 ùë°}\tag{53}$$
By integrating (53), we have:
$$ùëò(ùë°)=\frac{(ùêæ_0‚àíùêæ^‚àó )}{ùëÅ} ùëí^{\lambda_2 ùë°}+ùê¥_3\tag{54}$$
Evaluating (54) at $t = 0$, we get that:
$$ùëò(0)=\frac{(ùêæ_0‚àíùêæ^‚àó )}{ùëÅ}+ùê¥_3‚áíùê¥_3=\frac{ùëÅùëò(0)‚àíùêæ_0+ùêæ^‚àó}{ùëÅ}\tag{55}$$
Now, since the $N$ firms are identical, we have that:
$$ùêæ_0=ùêæ(0)=ùëÅùëò(0)\tag{56}  $$
Replacing (56) in (55), we get:
$$ùê¥_3=\frac{ùêæ^‚àó}{ùëÅ}\tag{57}$$
Evaluating (57) in (54), we have the optimal capital of the firm at each instant $t$. That is:
$$ùëò(t)=\frac{(ùêæ_0‚àíùêæ^‚àó )}{ùëÅ}ùëí^{\lambda_2 ùë°}+\frac{ùêæ^‚àó}{ùëÅ}\tag{58}$$
 <p style="text-align: justify;">Now, we will determine the equation of the $SB$ in the phase plane $(K, q)$. To do this, we must take into account that since the system (40) is linear, due to the functional forms adopted in (35), the $SB$ will be a straight line that passes through the stationary equilibrium point of the system (40) and has a slope equal to the slope of the eigenvector associated with the negative eigenvalue ($\lambda_2$). Therefore, the equation of the $SB$ will be:</p>
$$ùëû=\left(\frac{1‚àí\alpha\lambda_2ùêæ^‚àó}{ùëÅ}\right)+\left(\frac{\alpha\lambda_2}{ùëÅ}\right)ùêæ \tag{59}  $$
 <p style="text-align: justify;">Now, we will determine the equation of the unestable branch ($UB$) in the phase plane $(K, q)$. To do this, we must take into account that since the system (40) is linear, due to the functional forms adopted in (35), the $UB$ will be a straight line that passes through the stationary equilibrium point of the system (40) and has a slope equal to the slope of the eigenvector associated with the positive eigenvalue ($\lambda_1$). Therefore, the equation of the $UB$ will be:</p>
$$ùëû=\left(\frac{1‚àí\alpha\lambda_1ùêæ^‚àó}{ùëÅ}\right)+\left(\frac{\alpha\lambda_1}{ùëÅ}\right)ùêæ \tag{60}  $$

<p style="text-align: justify;">We will now carry out a numerical simulation of Tobin's q model for the following values of the model parameters:</p>

|$$ \alpha $$ | $$ a $$ | $$ b $$ | $$ r $$     | $$ N $$  | $$ K_0 $$  |
|-------------|---------|---------|-------------|----------|------------|

<p style="text-align: justify;">Replacing the parameters from the above table into equation (40) gives the ODE system to be simulated numerically:</p>
\begin{equation}
\left\{
\begin{aligned}
&\dot{K}(t)=0K(t)+1.25 ùëû(ùë°)‚àí1.25\\
&\dot{q}(t)=5ùêæ(ùë°)+0.3ùëû(ùë°)‚àí120\\
\end{aligned}
\right.
\tag{61}\end{equation}



  <p style="text-align: justify;">Substituting the values of the parameters in the above table into (42) and (45), respectively, it follows that:</p>

## Tobin_Solution_Final.m

```matlab

figure (1)

[t,x]=ode78('Tobin',0:0.1:9,[20, 1-3.152*(0.15-sqrt(6.2725))]);
K=x(:,1); % The first column is x(1) versus the time points at t.
q=x(:,2); % The second column is x(2)
plot(t,K,'r',t,q,'b');grid on;
xlabel('t (periods)')
ylabel('K(t) & q(t)')
title('Tobin Model Solutions')
legend('K(t)','q(t)')

figure(2)

[K,q] = meshgrid(0:0.009:30);
dK = 0*K+1.25*q-1.25;
dq = 5*K+0.3*q-120;
r = ( (dK).^2 + (dq).^2 ).^0.5;
pK = (dK)./r;
pq = (dq)./r;
streamslice(K,q,px,py);
xlabel('k(t)')
ylabel('q(t)')
title('Tobin Model Phase Diagram')
legend ('Phase Curves')

hold on

% Stable and Unestable Arms

w=linspace(0,30);
y1=(1+2.3544959572736387230121453634409*20*23.94/25)+(-2.3544959572736387230121453634409*20/25)*w;
y2=(1-2.6544959572736387230121453634409*20*23.94/25)+(2.6544959572736387230121453634409*20/25)*w;
plot(w,y1,'DisplayName','Stable arm')
plot(w,y2,'DisplayName','Unestable arm')

% Nullclines

x=linspace(0,30);
z1=1+0*x;
z2=-(50/3)*x+400;
plot(x,z1,'DisplayName','Nullcline: q = 1',color=[0,0,0],LineStyle='--')
plot(x,z2,'DisplayName','Nullcline: q = 400 - (50/3)*K',color=[0,0,0],LineStyle='-.')


legend;
axis([19 29 0 12])

hold off
```
